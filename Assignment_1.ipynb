# Q1) The Probability of Shooting at a Target

Given:

- Radius of disk = 10
- Event \(A\): the point lies within 1 unit of the boundary.

Therefore,
$$
\
A=\{(x,y): 9 \le r \le 10\}
\
$$
---

## 1) Uniform Distribution with Respect to Area

Total area of the disk:

$$
\pi(10)^2 = 100\pi
$$

Area of inner disk of radius 9:

$$
\pi(9)^2 = 81\pi
$$

Area within 1 unit of the boundary:

$$
100\pi - 81\pi = 19\pi
$$

Hence,

$$
P(A)=\frac{19\pi}{100\pi}
$$

$$
P(A)=\frac{19}{100}=0.19
$$

---

## 2) Radius Chosen Uniformly from \([0,10]\)

Here the radius \(r\) is uniformly distributed on \([0,10]\).

Event \(A\):

$$
9 \le r \le 10
$$

Using the uniform distribution formula:

$$
P(A)=\frac{\text{Length of favorable interval}}{\text{Total interval length}}
$$

$$
P(A)=\frac{10-9}{10-0}
$$

$$
P(A)=\frac{1}{10}=0.1
$$

---

## 3) Why Are the Answers Different?

### Model 1:
Probability is proportional to area.

The outer ring occupies a larger area, so it receives a larger probability.

### Model 2:
The radius itself is chosen uniformly.

Thus every radial interval of equal length has equal probability.

For example:

$$
[0,1], [5,6], [9,10]
$$

all have probability:

$$
0.1
$$

even though the outer rings contain larger physical area.

Therefore:

$$
P(A)=0.19 \quad \text{(uniform area model)}
$$

$$
P(A)=0.10 \quad \text{(uniform radius model)}
$$

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.”  
We need to find the probability that its length is greater than the side length of the inscribed equilateral triangle.

Let the circle have radius \(R\).

The side length of the inscribed equilateral triangle is:

$$
\sqrt{3}R
$$

A chord is longer than this if its perpendicular distance from the center is less than:

$$
\frac{R}{2}
$$

So the required condition is:

$$
d < \frac{R}{2}
$$

---

## Model 1: Random Distance from the Center

Choose the perpendicular distance \(d\) uniformly from:

$$
[0,R]
$$

Favorable case:

$$
0 \le d < \frac{R}{2}
$$

Therefore,

$$
P=\frac{R/2}{R}
$$

$$
P=\frac{1}{2}
$$

---

## Model 2: Random Midpoint in the Disk

Choose the midpoint of the chord uniformly inside the circle.

A chord is longer than $(\sqrt{3}R)$ if its midpoint lies inside a smaller circle of radius:

$$
\frac{R}{2}
$$

Area of smaller circle:

$$
\pi\left(\frac{R}{2}\right)^2
=
\frac{\pi R^2}{4}
$$

Area of full circle:

$$
\pi R^2
$$

Therefore,

$$
P=\frac{\pi R^2/4}{\pi R^2}
$$

$$
P=\frac{1}{4}
$$

---

## Model 3: Random Endpoint on the Circumference

Fix one endpoint of the chord.  
Choose the other endpoint uniformly on the circumference.

The chord is longer than the side of the inscribed equilateral triangle when the second point lies on the opposite one-third arc of the circle.

Therefore,

$$
P=\frac{1}{3}
$$

---

## Final Answers

Different natural ways of choosing a random chord give different probabilities:

$$
\boxed{\frac{1}{2}}
$$

$$
\boxed{\frac{1}{4}}
$$

$$
\boxed{\frac{1}{3}}
$$

---

## Explanation

The phrase “choose a chord at random” is not precise enough.

Different random selection methods create different probability spaces.

Therefore, Bertrand's paradox shows that a probability problem must clearly define the random experiment before calculating the probability.

# Q8) Probability of Being a Girl?

A family has two children.

$$
\{BG,GB,GG\}

P(GG \mid \text{at least one girl})

$$
GG

$$
BG,GB,GG
$$

So,

$$
P(GG \mid \text{at least one girl})

\frac{1}{3}
$$

$$
\boxed{\frac{1}{3}}

# Q9) Information Available in a Discrete Random Variable

## Given Sample Space

$$
\Omega=\{G_1,G_2,D_1,D_2\}
$$

with probabilities:

$$
P(G_1)=0.50,\quad P(G_2)=0.20,\quad P(D_1)=0.10,\quad P(D_2)=0.20
$$

---

## 1) Probability Space

Since the sample space is finite,

$$
\mathcal{F}=\mathcal{P}(\Omega)
$$

Therefore,

$$
(\Omega,\mathcal{F},P)
$$

where

$$
\mathcal{F}=\mathcal{P}(\Omega)
$$

---

## 2) Sigma-Algebra Generated by \(X\)

The random variable \(X\) records only whether the item is good or defective:

$$
X(G_1)=0,\quad X(G_2)=0
$$

$$
X(D_1)=1,\quad X(D_2)=1
$$

So \(X\) partitions the sample space into:

$$
\{G_1,G_2\}
$$

and

$$
\{D_1,D_2\}
$$

Therefore,

$$
\sigma(X)=\{\emptyset,\{G_1,G_2\},\{D_1,D_2\},\Omega\}
$$

---

## 3) Sigma-Algebra Generated by \(Y\)

The random variable \(Y\) records both quality and machine:

$$
Y(G_1)=1,\quad Y(G_2)=2,\quad Y(D_1)=3,\quad Y(D_2)=4
$$

Each outcome has a different value.

Therefore,

$$
\sigma(Y)=\mathcal{P}(\Omega)
$$

That is, \(Y\) can distinguish every individual outcome.

---


## 4) Show that $(\sigma(X)\subset\sigma(Y))$, and explain why \(Y\) contains more information than \(X\)

From Question 2,

$$
\sigma(X)=\{\emptyset,\{G_1,G_2\},\{D_1,D_2\},\Omega\}
$$

From Question 3,

$$
\sigma(Y)=\mathcal{P}(\Omega)
$$

Since $(\sigma(Y))$ is the power set of $(\Omega)$, it contains every subset of $(\Omega).$

Therefore, every set in $(\sigma(X))$ is also contained in $(\sigma(Y)).$

Hence,

$$
\boxed{\sigma(X)\subset\sigma(Y)}
$$

---

### Why does \(Y\) contain more information than \(X\)?

The random variable \(X\) only tells whether the product is:

- good, or
- defective.

It does not tell which machine produced the product.

However, \(Y\) records:

- product quality, and
- machine number.

Thus, \(Y\) can distinguish all individual outcomes:

$$
G_1,\;G_2,\;D_1,\;D_2
$$

while \(X\) only distinguishes:

$$
\{G_1,G_2\}
\quad \text{and} \quad
\{D_1,D_2\}
$$

Therefore, \(Y\) contains more detailed information than \(X\).

---

## 5) Marginal Distribution of \(X\)

$$
P(X=0)=P(G_1)+P(G_2)
$$

$$
P(X=0)=0.50+0.20=0.70
$$

For \(X=1\):

$$
P(X=1)=P(D_1)+P(D_2)
$$

$$

$$

Therefore,

$$
\boxed{P(X=0)=0.70,\quad P(X=1)=0.30}
$$

$$

Therefore,

$$
\boxed{
P(Y=1)=0.50,\quad
P(Y=2)=0.20,\quad
P(Y=3)=0.10,\quad
P(Y=4)=0.20
}
$$

---

## 7) Joint Probabilities

$$

$$

$$

$$

---

### (b) $P(X=1 \mid Y=4)$

If

$$
Y=4
$$

then the outcome must be

$$
D_2
$$

Since $D_2$ is defective,

$$
X=1
$$

Therefore,

$$
P(X=1 \mid Y=4)=1
$$

---

### (c) $P(Y=3 \mid X=1)$

Using the conditional probability formula:

$$
P(Y=3 \mid X=1)
=
\frac{P(Y=3,X=1)}{P(X=1)}
$$

Now,

$$
P(Y=3,X=1)=P(D_1)=0.10
$$

and

$$
P(X=1)=P(D_1)+P(D_2)
$$

$$
=0.10+0.20
$$

$$
=0.30
$$

Substituting,

$$
P(Y=3 \mid X=1)
=
\frac{0.10}{0.30}
$$

$$
=
\frac{1}{3}
$$

---

### (d) $P(Y=4 \mid X=1)$

Similarly,

$$
P(Y=4 \mid X=1)
=
\frac{P(Y=4,X=1)}{P(X=1)}
$$

Now,

$$
P(Y=4,X=1)=P(D_2)=0.20
$$

and

$$
P(X=1)=0.30
$$

Therefore,

$$
P(Y=4 \mid X=1)
=
\frac{0.20}{0.30}
$$

$$
=
\frac{2}{3}
$$

## Final Answers

$$

$$

$$

Conditioning on \(Y\) gives complete category information.

For example:

Thus, conditioning on \(Y\) provides more information than conditioning on \(X\).

# Q10) Information Available in Continuous Random Variables

## Given

Let

- \(T\) = reactor temperature deviation
- \(P\) = reactor pressure deviation

Assume:

$$
T \sim N(0,1)
$$

$$
P \sim N(0,4)
$$

and \(T\) and \(P\) are independent.

The probability space is:

$$
\Omega=\mathbb{R}^2
$$

$$
\mathcal{F}=\mathcal{B}(\mathbb{R}^2)
$$

with joint density:

$$
f_{T,P}(t,p)
=
\frac{1}{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
$$

Define:

$$
X(t,p)=t
$$

$$
Y(t,p)=(t,p)
$$

---

# 1) Probability Space

The probability space is:

$$
(\Omega,\mathcal{F},P)
$$

where

$$
\Omega=\mathbb{R}^2
$$

$$
\mathcal{F}=\mathcal{B}(\mathbb{R}^2)
$$

and \(P\) is induced by the joint density

$$
f_{T,P}(t,p)
=
\frac{1}{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
$$

---

# 2) Sigma-Algebra Generated by \(X\)

The random variable \(X\) records only temperature deviation.

Thus, $(\sigma(X))$ contains events depending only on \(t\).

Examples:

$$
\{(t,p): t\le 0\}
$$

$$
\{(t,p): -1<t<1\}
$$

Therefore,

$$
\sigma(X)
=
\{
X^{-1}(B): B\in\mathcal{B}(\mathbb{R})
\}
$$

---

# 3) Sigma-Algebra Generated by \(Y\)

The random variable \(Y\) records both temperature and pressure.

Thus,

$$
Y(t,p)=(t,p)
$$

contains complete information about the state.

Therefore,

$$
\sigma(Y)=\mathcal{B}(\mathbb{R}^2)
$$

---

## 4) Show that $\sigma(X)\subseteq\sigma(Y)$, and explain why $Y$ contains more information than $X$

The random variable $X$ is defined as:

$$
X(t,p)=t
$$

Thus, $X$ records only the reactor temperature deviation.

The random variable $Y$ is defined as:

$$
Y(t,p)=(t,p)
$$

Thus, $Y$ records both:

- temperature deviation $t$
- pressure deviation $p$

Every event determined by $X$ depends only on the temperature coordinate $t$.

Since $Y$ contains both coordinates $(t,p)$, any event measurable with respect to $X$ can also be determined using $Y$.

Therefore,

$$
\boxed{\sigma(X)\subseteq\sigma(Y)}
$$



### Why does $Y$ contain more information than $X$?

The variable $X$ only provides information about temperature deviation.

For example, using $X$, we can determine events such as:

$$
\{(t,p): t<0\}
$$

However, $X$ cannot determine events involving pressure alone, such as:

$$
\{(t,p): p>2\}
$$

The variable $Y$ records the complete pair:

$$
(t,p)
$$

Therefore, $Y$ can determine:

- temperature-based events,
- pressure-based events,
- and combined events involving both variables.

Hence, $Y$ contains more detailed information than $X$.

---

## 5) Marginal Density of $X$

The random variable $X$ is defined as:

$$
X(t,p)=t
$$

Thus, $X$ represents only the temperature deviation.

The joint density of $(T,P)$ is:

$$
f_{T,P}(t,p)
=
\frac{1}{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
$$

To find the marginal density of $X$, we integrate the joint density over all possible values of $p$:

$$
f_X(t)
=
\int_{-\infty}^{\infty}
f_{T,P}(t,p)\,dp
$$

Substituting the joint density:

$$
f_X(t)
=
\int_{-\infty}^{\infty}
\frac{1}{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
dp
$$

Since $t$ is constant with respect to $p$:

$$
f_X(t)
=
\frac{1}{4\pi}
e^{-t^2/2}
\int_{-\infty}^{\infty}
e^{-p^2/8}\,dp
$$

Using the Gaussian integral result:

$$
\int_{-\infty}^{\infty}
e^{-p^2/8}\,dp
=
2\sqrt{2\pi}
$$

Therefore,

$$
f_X(t)
=
\frac{1}{\sqrt{2\pi}}
e^{-t^2/2}
$$

Hence, the marginal density of $X$ is:

$$
\boxed{
f_X(t)
=
\frac{1}{\sqrt{2\pi}}
e^{-t^2/2}
}
$$

Thus,

$$
X\sim N(0,1)
$$

---

## 6) Marginal Density of $Y$

The random variable $Y$ is defined as:

$$
Y(t,p)=(t,p)
$$

Thus, $Y$ records both:

- temperature deviation $t$
- pressure deviation $p$

The joint density of $(T,P)$ is:

$$
f_{T,P}(t,p)
=
\frac{1}{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
$$

Since

$$
Y=(T,P)
$$

the marginal density of $Y$ is the same as the joint density of $(T,P)$.

Therefore,

$$
f_Y(t,p)
=
\frac{1}{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
$$

Hence, the marginal density of $Y$ is:

$$
\boxed{
f_Y(t,p)
=
\frac{1}{4\pi}
\exp\left(
-\frac{t^2}{2}
-\frac{p^2}{8}
\right)
}
$$

This density describes the probability distribution of the complete reactor state:

$$
(t,p)
$$

including both temperature and pressure deviations.

---

## 7) Compute the Required Probabilities

### (a) $P(X\le 0)$

Since $X\sim N(0,1)$, the standard normal distribution is symmetric about $0$.

Therefore,

$$
P(X\le 0)=\frac{1}{2}
$$

Hence,

$$
\boxed{P(X\le 0)=0.5}
$$

---

### (b) $P(X>1)$

Using the standard normal CDF $\Phi$:

$$
P(X>1)=1-\Phi(1)
$$

From the standard normal table,

$$
\Phi(1)\approx 0.8413
$$

Therefore,

$$
P(X>1)=1-0.8413
$$

$$
P(X>1)=0.1587
$$

Hence,

$$
\boxed{P(X>1)=0.1587}
$$

---

### (c) $P\left(Y\in(-\infty,0]\times\mathbb{R}\right)$

Since

$$
Y=(T,P)
$$

the event

$$
Y\in(-\infty,0]\times\mathbb{R}
$$

means:

$$
T\le 0
$$

Since $T\sim N(0,1)$,

$$
P(T\le 0)=0.5
$$

Therefore,

$$
\boxed{
P\left(Y\in(-\infty,0]\times\mathbb{R}\right)=0.5
}
$$

---

### (d) $P\left(Y\in[-1,1]\times[-2,2]\right)$

This event means:

$$
-1\le T\le 1
$$

and

$$
-2\le P\le 2
$$

So,

$$
P\left(Y\in[-1,1]\times[-2,2]\right)
=
P(-1\le T\le 1,\,-2\le P\le 2)
$$

Since $T$ and $P$ are independent,

$$
P(-1\le T\le 1,\,-2\le P\le 2)
=
P(-1\le T\le 1)\cdot P(-2\le P\le 2)
$$

For $T\sim N(0,1)$,

$$
P(-1\le T\le 1)\approx 0.6827
$$

For $P\sim N(0,4)$, the standard deviation is:

$$
\sigma=2
$$

Standardizing,

$$
P(-2\le P\le 2)
=
P\left(\frac{-2}{2}\le Z\le \frac{2}{2}\right)
$$

$$
=
P(-1\le Z\le 1)
$$

$$
\approx 0.6827
$$

Therefore,

$$
P\left(Y\in[-1,1]\times[-2,2]\right)
=
0.6827\times 0.6827
$$

$$
\approx 0.4661
$$

Hence,

$$
\boxed{
P\left(Y\in[-1,1]\times[-2,2]\right)\approx 0.4661
}
$$

---

## 8) Conditional Distribution of $X\mid Y=(t,p)$

The random variable $Y$ is defined as:

$$
Y(t,p)=(t,p)
$$

So if we are given:

$$
Y=(t,p)
$$

then both temperature and pressure are known exactly.

The random variable $X$ is defined as:

$$
X(t,p)=t
$$

Therefore, once $Y=(t,p)$ is known, the value of $X$ is also known exactly:

$$
X=t
$$

Hence, the conditional distribution of $X\mid Y=(t,p)$ is a degenerate distribution at $t$.

This means:

$$
\boxed{P(X=t\mid Y=(t,p))=1}
$$

In words, if the complete state $(t,p)$ is known, then the temperature value $t$ is already known.

---

## 9) Conditional Distribution of $Y\mid X=t$

The random variable $X$ is defined as:

$$
X(t,p)=t
$$

So if we are given:

$$
X=t
$$

then the temperature deviation is known.

However, the pressure deviation $P$ is still unknown.

Since $T$ and $P$ are independent, knowing $T=t$ does not change the distribution of $P$.

Therefore,

$$
P\mid X=t \sim N(0,4)
$$

Since

$$
Y=(T,P)
$$

we get:

$$
Y\mid X=t=(t,P)
$$

where

$$
P\sim N(0,4)
$$

Hence,

$$
\boxed{
Y\mid X=t=(t,P),\quad P\sim N(0,4)
}
$$

In words, conditioning on $X=t$ fixes the temperature value, but the pressure remains random.

---

## 10) Information in Conditioning on $X$ and $Y$

Conditioning on $X$ means we know only the temperature deviation.

Since

$$
X(t,p)=t
$$

knowing $X$ tells us the value of $t$, but not the value of $p$.

So, when we condition on $X$:

- temperature is known
- pressure is still unknown

---

Conditioning on $Y$ means we know the complete reactor state.

Since

$$
Y(t,p)=(t,p)
$$

knowing $Y$ tells us both $t$ and $p$.

So, when we condition on $Y$:

- temperature is known
- pressure is known

---

Therefore, conditioning on $Y$ gives more information than conditioning on $X$.

This is because

$$
\sigma(X)\subseteq\sigma(Y)
$$

Hence,

$$
\boxed{
\text{Conditioning on }Y\text{ is more informative than conditioning on }X.
}
$$

## 11) Show that $X=\pi_1\circ Y$

Define the projection function $\pi_1$ as:

$$
\pi_1(t,p)=t
$$

This function takes the first coordinate of the pair $(t,p)$.

Now, the random variable $Y$ is:

$$
Y(t,p)=(t,p)
$$

Apply $\pi_1$ to $Y(t,p)$:

$$
(\pi_1\circ Y)(t,p)
=
\pi_1(Y(t,p))
$$

$$
=
\pi_1(t,p)
$$

$$
=
t
$$

But,

$$
X(t,p)=t
$$

Therefore,

$$
(\pi_1\circ Y)(t,p)=X(t,p)
$$

Hence,

$$
\boxed{
X=\pi_1\circ Y
}
$$

This shows that $X$ can be obtained from $Y$ by taking the first coordinate.

Therefore, $X$ is measurable with respect to $Y$, and this implies:

$$
\boxed{
\sigma(X)\subseteq\sigma(Y)
}
$$
