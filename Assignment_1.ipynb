# Q1) The Probability of Shooting at a Target

Given:

- Radius of disk = 10
- Event \(A\): the point lies within 1 unit of the boundary.

Therefore,
$$
\
A=\{(x,y): 9 \le r \le 10\}
\
$$
---

## 1) Uniform Distribution with Respect to Area

Total area of the disk:

$$
\pi(10)^2 = 100\pi
$$

Area of inner disk of radius 9:

$$
\pi(9)^2 = 81\pi
$$

Area within 1 unit of the boundary:

$$
100\pi - 81\pi = 19\pi
$$

Hence,

$$
P(A)=\frac{19\pi}{100\pi}
$$

$$
P(A)=\frac{19}{100}=0.19
$$

---

## 2) Radius Chosen Uniformly from \([0,10]\)

Here the radius \(r\) is uniformly distributed on \([0,10]\).

Event \(A\):

$$
9 \le r \le 10
$$

Using the uniform distribution formula:

$$
P(A)=\frac{\text{Length of favorable interval}}{\text{Total interval length}}
$$

$$
P(A)=\frac{10-9}{10-0}
$$

$$
P(A)=\frac{1}{10}=0.1
$$

---

## 3) Why Are the Answers Different?

### Model 1:
Probability is proportional to area.

The outer ring occupies a larger area, so it receives a larger probability.

### Model 2:
The radius itself is chosen uniformly.

Thus every radial interval of equal length has equal probability.

For example:

$$
[0,1], [5,6], [9,10]
$$

all have probability:

$$
0.1
$$

even though the outer rings contain larger physical area.

Therefore:

$$
P(A)=0.19 \quad \text{(uniform area model)}
$$

$$
P(A)=0.10 \quad \text{(uniform radius model)}
$$

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.”  
We need to find the probability that its length is greater than the side length of the inscribed equilateral triangle.

Let the circle have radius \(R\).

The side length of the inscribed equilateral triangle is:

$$
\sqrt{3}R
$$

A chord is longer than this if its perpendicular distance from the center is less than:

$$
\frac{R}{2}
$$

So the required condition is:

$$
d < \frac{R}{2}
$$

---

## Model 1: Random Distance from the Center

Choose the perpendicular distance \(d\) uniformly from:

$$
[0,R]
$$

Favorable case:

$$
0 \le d < \frac{R}{2}
$$

Therefore,

$$
P=\frac{R/2}{R}
$$

$$
P=\frac{1}{2}
$$

---

## Model 2: Random Midpoint in the Disk

Choose the midpoint of the chord uniformly inside the circle.

A chord is longer than $(\sqrt{3}R)$ if its midpoint lies inside a smaller circle of radius:

$$
\frac{R}{2}
$$

Area of smaller circle:

$$
\pi\left(\frac{R}{2}\right)^2
=
\frac{\pi R^2}{4}
$$

Area of full circle:

$$
\pi R^2
$$

Therefore,

$$
P=\frac{\pi R^2/4}{\pi R^2}
$$

$$
P=\frac{1}{4}
$$

---

## Model 3: Random Endpoint on the Circumference

Fix one endpoint of the chord.  
Choose the other endpoint uniformly on the circumference.

The chord is longer than the side of the inscribed equilateral triangle when the second point lies on the opposite one-third arc of the circle.

Therefore,

$$
P=\frac{1}{3}
$$

---

## Final Answers

Different natural ways of choosing a random chord give different probabilities:

$$
\boxed{\frac{1}{2}}
$$

$$
\boxed{\frac{1}{4}}
$$

$$
\boxed{\frac{1}{3}}
$$

---

## Explanation

The phrase “choose a chord at random” is not precise enough.

Different random selection methods create different probability spaces.

Therefore, Bertrand's paradox shows that a probability problem must clearly define the random experiment before calculating the probability.

# Q8) Probability of Being a Girl?

A family has two children.

Each child can be:

$$
B = \text{Boy}, \quad G = \text{Girl}
$$

The sample space is:

$$
\Omega=\{BB,BG,GB,GG\}
$$

Each outcome has equal probability:

$$
P(BB)=P(BG)=P(GB)=P(GG)=\frac{1}{4}
$$

We are given that at least one child is a girl.

So remove \(BB\).

Remaining possible outcomes:

$$
\{BG,GB,GG\}
$$

We want the probability that both children are girls:

$$
P(GG \mid \text{at least one girl})
$$

Only one outcome is favorable:

$$
GG
$$

Total possible outcomes after the condition:

$$
BG,GB,GG
$$

So,

$$
P(GG \mid \text{at least one girl})
=
\frac{1}{3}
$$

Therefore,

$$
\boxed{\frac{1}{3}}
$$

# Q9) Information Available in a Discrete Random Variable

## Given Sample Space

$$
\Omega=\{G_1,G_2,D_1,D_2\}
$$

with probabilities:

$$
P(G_1)=0.50,\quad P(G_2)=0.20,\quad P(D_1)=0.10,\quad P(D_2)=0.20
$$

---

## 1) Probability Space

Since the sample space is finite,

$$
\mathcal{F}=\mathcal{P}(\Omega)
$$

Therefore,

$$
(\Omega,\mathcal{F},P)
$$

where

$$
\mathcal{F}=\mathcal{P}(\Omega)
$$

---

## 2) Sigma-Algebra Generated by \(X\)

The random variable \(X\) records only whether the item is good or defective:

$$
X(G_1)=0,\quad X(G_2)=0
$$

$$
X(D_1)=1,\quad X(D_2)=1
$$

So \(X\) partitions the sample space into:

$$
\{G_1,G_2\}
$$

and

$$
\{D_1,D_2\}
$$

Therefore,

$$
\sigma(X)=\{\emptyset,\{G_1,G_2\},\{D_1,D_2\},\Omega\}
$$

---

## 3) Sigma-Algebra Generated by \(Y\)

The random variable \(Y\) records both quality and machine:

$$
Y(G_1)=1,\quad Y(G_2)=2,\quad Y(D_1)=3,\quad Y(D_2)=4
$$

Each outcome has a different value.

Therefore,

$$
\sigma(Y)=\mathcal{P}(\Omega)
$$

That is, \(Y\) can distinguish every individual outcome.

---


## 4) Show that $(\sigma(X)\subset\sigma(Y))$, and explain why \(Y\) contains more information than \(X\)

From Question 2,

$$
\sigma(X)=\{\emptyset,\{G_1,G_2\},\{D_1,D_2\},\Omega\}
$$

From Question 3,

$$
\sigma(Y)=\mathcal{P}(\Omega)
$$

Since $(\sigma(Y))$ is the power set of $(\Omega)$, it contains every subset of $(\Omega).$

Therefore, every set in $(\sigma(X))$ is also contained in $(\sigma(Y)).$

Hence,

$$
\boxed{\sigma(X)\subset\sigma(Y)}
$$

---

### Why does \(Y\) contain more information than \(X\)?

The random variable \(X\) only tells whether the product is:

- good, or
- defective.

It does not tell which machine produced the product.

However, \(Y\) records:

- product quality, and
- machine number.

Thus, \(Y\) can distinguish all individual outcomes:

$$
G_1,\;G_2,\;D_1,\;D_2
$$

while \(X\) only distinguishes:

$$
\{G_1,G_2\}
\quad \text{and} \quad
\{D_1,D_2\}
$$

Therefore, \(Y\) contains more detailed information than \(X\).

---

## 5) Marginal Distribution of \(X\)

For \(X=0\):

$$
P(X=0)=P(G_1)+P(G_2)
$$

$$
P(X=0)=0.50+0.20=0.70
$$

For \(X=1\):

$$
P(X=1)=P(D_1)+P(D_2)
$$

$$
P(X=1)=0.10+0.20=0.30
$$

Therefore,

$$
\boxed{P(X=0)=0.70,\quad P(X=1)=0.30}
$$

---

## 6) Marginal Distribution of \(Y\)

$$
P(Y=1)=P(G_1)=0.50
$$

$$
P(Y=2)=P(G_2)=0.20
$$

$$
P(Y=3)=P(D_1)=0.10
$$

$$
P(Y=4)=P(D_2)=0.20
$$

Therefore,

$$
\boxed{
P(Y=1)=0.50,\quad
P(Y=2)=0.20,\quad
P(Y=3)=0.10,\quad
P(Y=4)=0.20
}
$$

---

## 7) Joint Probabilities

$$
P(X=0,Y=1)=P(G_1)=0.50
$$

$$
P(X=0,Y=2)=P(G_2)=0.20
$$

$$
P(X=1,Y=3)=P(D_1)=0.10
$$

$$
P(X=1,Y=4)=P(D_2)=0.20
$$

---

## 8) Conditional Probabilities

#### (a) $P(X=1 \mid Y=3)$

If

$$
Y=3
$$

then the outcome must be

$$
D_1
$$

Since $D_1$ is defective,

$$
X=1
$$

Therefore,

$$
P(X=1 \mid Y=3)=1
$$

---

### (b) $P(X=1 \mid Y=4)$

If

$$
Y=4
$$

then the outcome must be

$$
D_2
$$

Since $D_2$ is defective,

$$
X=1
$$

Therefore,

$$
P(X=1 \mid Y=4)=1
$$

---

### (c) $P(Y=3 \mid X=1)$

Using the conditional probability formula:

$$
P(Y=3 \mid X=1)
=
\frac{P(Y=3,X=1)}{P(X=1)}
$$

Now,

$$
P(Y=3,X=1)=P(D_1)=0.10
$$

and

$$
P(X=1)=P(D_1)+P(D_2)
$$

$$
=0.10+0.20
$$

$$
=0.30
$$

Substituting,

$$
P(Y=3 \mid X=1)
=
\frac{0.10}{0.30}
$$

$$
=
\frac{1}{3}
$$

---

### (d) $P(Y=4 \mid X=1)$

Similarly,

$$
P(Y=4 \mid X=1)
=
\frac{P(Y=4,X=1)}{P(X=1)}
$$

Now,

$$
P(Y=4,X=1)=P(D_2)=0.20
$$

and

$$
P(X=1)=0.30
$$

Therefore,

$$
P(Y=4 \mid X=1)
=
\frac{0.20}{0.30}
$$

$$
=
\frac{2}{3}
$$

---

## Final Answers

$$
\boxed{P(X=1 \mid Y=3)=1}
$$

$$
\boxed{P(X=1 \mid Y=4)=1}
$$

$$
\boxed{P(Y=3 \mid X=1)=\frac{1}{3}}
$$

$$
\boxed{P(Y=4 \mid X=1)=\frac{2}{3}}
$$

---

## 9) Difference Between Conditioning on \(X\) and Conditioning on \(Y\)

Conditioning on \(X\) gives only quality information.

For example:

$$
X=1
$$

means the product is defective, but we do not know which machine produced it.

Conditioning on \(Y\) gives complete category information.

For example:

$$
Y=3
$$

means the product is defective and produced by Machine 1.

Thus, conditioning on \(Y\) provides more information than conditioning on \(X\).

---

## 10) Information in $\sigma(X)$ and $\sigma(Y)$

There is no information in $\sigma(X)$ that is missing from $\sigma(Y)$, because

$$
\sigma(X)\subset\sigma(Y)
$$

However, $\sigma(Y)$ contains additional information such as:

- whether the product came from Machine 1 or Machine 2
- the exact category:
  
$$
G_1,\;G_2,\;D_1,\;D_2
$$

The random variable $X$ only tells whether the product is:

- good, or
- defective.

But the random variable $Y$ tells:

- product quality, and
- machine number.

Therefore, $Y$ provides more detailed information than $X$.

```python
## 11) Python-Based Visualization of the Full Model


import matplotlib.pyplot as plt
import pandas as pd

# -----------------------------
# Data
# -----------------------------
data = pd.DataFrame({
    "Outcome": ["G1", "G2", "D1", "D2"],
    "Meaning": [
        "Good, Machine 1",
        "Good, Machine 2",
        "Defective, Machine 1",
        "Defective, Machine 2"
    ],
    "Probability": [0.50, 0.20, 0.10, 0.20],
    "X": [0, 0, 1, 1],
    "Y": [1, 2, 3, 4]
})

display(data)

# -----------------------------
# 1) Probability distribution of outcomes
# -----------------------------
plt.figure(figsize=(7,4))
plt.bar(data["Outcome"], data["Probability"])
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Probability Distribution on Ω")
plt.ylim(0, 0.6)

for i, p in enumerate(data["Probability"]):
    plt.text(i, p + 0.02, str(p), ha="center")

plt.show()

# -----------------------------
# 2) Marginal distribution of X
# -----------------------------
x_dist = data.groupby("X")["Probability"].sum()

plt.figure(figsize=(6,4))
plt.bar(["X=0 (Good)", "X=1 (Defective)"], x_dist.values)
plt.ylabel("Probability")
plt.title("Marginal Distribution of X")
plt.ylim(0, 0.8)

for i, p in enumerate(x_dist.values):
    plt.text(i, p + 0.02, str(round(p,2)), ha="center")

plt.show()

# -----------------------------
# 3) Marginal distribution of Y
# -----------------------------
y_dist = data.groupby("Y")["Probability"].sum()

plt.figure(figsize=(7,4))
plt.bar(["Y=1", "Y=2", "Y=3", "Y=4"], y_dist.values)
plt.ylabel("Probability")
plt.title("Marginal Distribution of Y")
plt.ylim(0, 0.6)

for i, p in enumerate(y_dist.values):
    plt.text(i, p + 0.02, str(round(p,2)), ha="center")

plt.show()

# -----------------------------
# 4) Joint distribution of X and Y
# -----------------------------
joint = data.pivot_table(
    values="Probability",
    index="X",
    columns="Y",
    aggfunc="sum",
    fill_value=0
)

display(joint)

plt.figure(figsize=(7,4))
plt.imshow(joint, aspect="auto")
plt.colorbar(label="Probability")
plt.xticks(range(len(joint.columns)), [f"Y={y}" for y in joint.columns])
plt.yticks(range(len(joint.index)), [f"X={x}" for x in joint.index])
plt.title("Joint Distribution of X and Y")

for i in range(len(joint.index)):
    for j in range(len(joint.columns)):
        plt.text(j, i, joint.iloc[i, j], ha="center", va="center")

plt.show()

# -----------------------------
# 5) Information comparison: X vs Y
# -----------------------------
fig, ax = plt.subplots(figsize=(9,4))

ax.text(0.15, 0.7, "X information", fontsize=14, weight="bold")
ax.text(0.15, 0.55, "{G1, G2} → X=0", fontsize=12)
ax.text(0.15, 0.40, "{D1, D2} → X=1", fontsize=12)

ax.text(0.60, 0.7, "Y information", fontsize=14, weight="bold")
ax.text(0.60, 0.55, "G1 → Y=1", fontsize=12)
ax.text(0.60, 0.43, "G2 → Y=2", fontsize=12)
ax.text(0.60, 0.31, "D1 → Y=3", fontsize=12)
ax.text(0.60, 0.19, "D2 → Y=4", fontsize=12)

ax.text(0.32, 0.05, "σ(X) ⊂ σ(Y): Y contains more information than X", fontsize=13)

ax.axis("off")
plt.title("Information Available in X and Y")
plt.show()
```
