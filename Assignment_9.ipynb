# Q. Gaussian Mixture Clustering as Conditional Updating

Consider observations

$$
x_1,x_2,\ldots,x_n\in\mathbb{R}^d
$$

and a latent cluster variable

$$
C_i\in\{1,\ldots,K\}.
$$

The prior probability that observation $x_i$ belongs to cluster $k$ is

$$
P(C_i=k)=\phi_k,
\qquad
\phi_k\geq 0,
\qquad
\sum_{k=1}^{K}\phi_k=1.
$$

Conditional on $C_i=k$, the observation follows

$$
X_i\mid C_i=k\sim\mathcal{N}(\mu_k,\Sigma_k).
$$


## 1. Deriving the Marginal Density

Using the law of total probability, the marginal density of $X_i$ is obtained by summing over every possible value of the latent cluster variable $C_i$:

$$
p(x_i)
=
\sum_{k=1}^{K}p(x_i,C_i=k).
$$

Using the product rule,

$$
p(x_i,C_i=k)
=
p(x_i\mid C_i=k)P(C_i=k).
$$

Therefore,

$$
p(x_i)
=
\sum_{k=1}^{K}
p(x_i\mid C_i=k)P(C_i=k).
$$

From the model assumptions,

$$
p(x_i\mid C_i=k)
=
\mathcal{N}(x_i\mid\mu_k,\Sigma_k)
$$

and

$$
P(C_i=k)=\phi_k.
$$

Hence,

$$
\boxed{
p(x_i)
=
\sum_{k=1}^{K}
\phi_k\mathcal{N}(x_i\mid\mu_k,\Sigma_k)
}
$$

where the multivariate Gaussian density is

$$
\mathcal{N}(x_i\mid\mu_k,\Sigma_k)
=
\frac{1}
{(2\pi)^{d/2}|\Sigma_k|^{1/2}}
\exp\left[
-\frac{1}{2}
(x_i-\mu_k)^T
\Sigma_k^{-1}
(x_i-\mu_k)
\right].
$$

This is called a **Gaussian mixture density** because it is a weighted combination, or mixture, of $K$ Gaussian component densities. Each component represents one cluster, and $\phi_k$ is the mixing weight of cluster $k$.

Since

$$
\phi_k\geq 0
\qquad\text{and}\qquad
\sum_{k=1}^{K}\phi_k=1,
$$

the marginal density is a valid probability density. It can represent distributions that are multimodal, asymmetric, or more complicated than a single Gaussian distribution.

## 2. Deriving the Posterior Cluster Probability

For a fixed observation $x_i$, Bayes' rule gives

$$
P(C_i=k\mid X_i=x_i)
=
\frac{
P(X_i=x_i\mid C_i=k)P(C_i=k)
}{
P(X_i=x_i)
}.
$$

The denominator is obtained by summing over all possible clusters:

$$
P(X_i=x_i)
=
\sum_{j=1}^{K}
P(X_i=x_i\mid C_i=j)P(C_i=j).
$$

Therefore,

$$
P(C_i=k\mid X_i=x_i)
=
\frac{
P(X_i=x_i\mid C_i=k)P(C_i=k)
}{
\sum_{j=1}^{K}
P(X_i=x_i\mid C_i=j)P(C_i=j)
}.
$$

Substituting the Gaussian likelihood and the cluster prior,

$$
P(X_i=x_i\mid C_i=k)
=
\mathcal{N}(x_i\mid\mu_k,\Sigma_k)
$$

and

$$
P(C_i=k)=\phi_k,
$$

we obtain

$$
\boxed{
P(C_i=k\mid X_i=x_i)
=
\frac{
\phi_k\mathcal{N}(x_i\mid\mu_k,\Sigma_k)
}{
\sum_{j=1}^{K}
\phi_j\mathcal{N}(x_i\mid\mu_j,\Sigma_j)
}
}
$$

This posterior probability is called the **responsibility** of cluster $k$ for observation $x_i$:

$$
\boxed{
\gamma_{ik}
=
P(C_i=k\mid X_i=x_i)
}
$$

The numerator

$$
\phi_k\mathcal{N}(x_i\mid\mu_k,\Sigma_k)
$$

combines:

- the prior probability $\phi_k$ that a point belongs to cluster $k$, and
- the likelihood $\mathcal{N}(x_i\mid\mu_k,\Sigma_k)$ of observing $x_i$ under cluster $k$.

The denominator

$$
\sum_{j=1}^{K}
\phi_j\mathcal{N}(x_i\mid\mu_j,\Sigma_j)
$$

normalizes the result across all clusters. Therefore,

$$
0\leq\gamma_{ik}\leq 1
$$

and

$$
\sum_{k=1}^{K}\gamma_{ik}=1.
$$

Thus, $\gamma_{ik}$ is interpreted as the updated probability that observation $x_i$ belongs to cluster $k$ after observing its value. It is a **soft cluster assignment**, because one observation may have nonzero membership probabilities for several clusters.

## 3. One-Hot Encoding of the Latent Cluster Variable

Define the one-hot encoded latent vector

$$
Z_i
=
\begin{bmatrix}
Z_{i1}\\
Z_{i2}\\
\vdots\\
Z_{iK}
\end{bmatrix},
$$

where

$$
Z_{ik}
=
\begin{cases}
1, & \text{if } C_i=k,\\
0, & \text{otherwise}.
\end{cases}
$$

Equivalently,

$$
Z_{ik}=\mathbf{1}_{\{C_i=k\}},
$$

where $\mathbf{1}_{\{C_i=k\}}$ is the indicator function of the event $C_i=k$.

The conditional expectation is

$$
\mathbb{E}[Z_{ik}\mid X_i=x_i]
=
1\cdot P(Z_{ik}=1\mid X_i=x_i)
+
0\cdot P(Z_{ik}=0\mid X_i=x_i).
$$

Because

$$
Z_{ik}=1
\quad\Longleftrightarrow\quad
C_i=k,
$$

we obtain

$$
\mathbb{E}[Z_{ik}\mid X_i=x_i]
=
P(C_i=k\mid X_i=x_i).
$$

Using the definition of responsibility,

$$
\boxed{
\mathbb{E}[Z_{ik}\mid X_i=x_i]
=
\gamma_{ik}
}
$$

for every $k=1,\ldots,K$.

Taking the conditional expectation component by component gives

$$
\mathbb{E}[Z_i\mid X_i=x_i]
=
\begin{bmatrix}
\mathbb{E}[Z_{i1}\mid X_i=x_i]\\
\mathbb{E}[Z_{i2}\mid X_i=x_i]\\
\vdots\\
\mathbb{E}[Z_{iK}\mid X_i=x_i]
\end{bmatrix}.
$$

Therefore,

$$
\boxed{
\mathbb{E}[Z_i\mid X_i=x_i]
=
\begin{bmatrix}
\gamma_{i1}\\
\gamma_{i2}\\
\vdots\\
\gamma_{iK}
\end{bmatrix}
}
$$

Hence, the soft cluster assignment in a Gaussian mixture model is precisely

$$
\boxed{
\mathbb{E}[Z_i\mid X_i=x_i]
}
$$

The latent vector $Z_i$ represents an unknown hard cluster assignment, while its conditional expectation gives the posterior probabilities of membership in all clusters. A hard assignment can be obtained, when required, by choosing

$$
\hat{C}_i
=
\arg\max_{k\in\{1,\ldots,K\}}\gamma_{ik}.
$$


## Final Conclusion

A Gaussian mixture model can be understood as a conditional updating problem:

1. Before observing $x_i$, the cluster membership probabilities are the prior probabilities

$$
P(C_i=k)=\phi_k.
$$

2. After observing $x_i$, Bayes' rule updates these probabilities to

$$
P(C_i=k\mid X_i=x_i)=\gamma_{ik}.
$$

3. The complete soft assignment vector is

$$
\mathbb{E}[Z_i\mid X_i=x_i]
=
\begin{bmatrix}
\gamma_{i1}\\
\gamma_{i2}\\
\vdots\\
\gamma_{iK}
\end{bmatrix}.
$$

Therefore, Gaussian mixture clustering performs probabilistic cluster assignment by updating prior cluster probabilities using the likelihood of the observed data point under each Gaussian component.
