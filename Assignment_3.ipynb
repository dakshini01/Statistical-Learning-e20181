#Q1) Prediction of fraudulent transactions
## 1) Probability Space

The sample space is

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\}.
$$

Since the question says

$$
\mathcal{F}=\mathcal{P}(\Omega),
$$

the sigma-algebra is the power set of $\Omega$:

$$
\mathcal{F}
=
\mathcal{P}(\Omega)
=
\{
\emptyset,
\{\omega_1\},
\{\omega_2\},
\{\omega_3\},
\{\omega_4\},
\{\omega_1,\omega_2\},
\{\omega_1,\omega_3\},
\{\omega_1,\omega_4\},
\{\omega_2,\omega_3\},
\{\omega_2,\omega_4\},
\{\omega_3,\omega_4\},
\{\omega_1,\omega_2,\omega_3\},
\{\omega_1,\omega_2,\omega_4\},
\{\omega_1,\omega_3,\omega_4\},
\{\omega_2,\omega_3,\omega_4\},
\Omega
\}.
$$

The probability measure $P$ is defined by

$$
P(\{\omega_1\})=0.50,
$$

$$
P(\{\omega_2\})=0.10,
$$

$$
P(\{\omega_3\})=0.20,
$$

$$
P(\{\omega_4\})=0.20.
$$

Therefore, the probability space is

$$
(\Omega,\mathcal{F},P).
$$

# 2) Sigma-Algebra Generated by $Y$

The random variable $Y$ is defined as

$$
Y(\omega_1)=0,\quad Y(\omega_2)=0,\quad Y(\omega_3)=1,\quad Y(\omega_4)=1.
$$

So $Y$ separates the sample space into two sets:

$$
\{Y=0\}=\{\omega_1,\omega_2\}
$$

and

$$
\{Y=1\}=\{\omega_3,\omega_4\}.
$$

Therefore, the sigma-algebra generated by $Y$ is

$$
\sigma(Y)
=
\{\emptyset,\Omega,\{\omega_1,\omega_2\},\{\omega_3,\omega_4\}\}.
$$

# 3) Why $E[X \mid Y]$ Must Be $\sigma(Y)$-Measurable

The conditional expectation $E[X \mid Y]$ is the best estimate of $X$ using only the information given by $Y$.

Since $Y$ only tells whether the customer is low-risk or high-risk, $E[X \mid Y]$ can only depend on the value of $Y$.

From Part 2,

$$
\sigma(Y)
=
\{\emptyset,\Omega,\{\omega_1,\omega_2\},\{\omega_3,\omega_4\}\}.
$$

This means the available information separates $\Omega$ into the two sets

$$
\{\omega_1,\omega_2\}
$$

and

$$
\{\omega_3,\omega_4\}.
$$

Therefore, any $\sigma(Y)$-measurable random variable must be constant on each of these sets.

Hence,

$$
E[X \mid Y]
$$

must be constant on

$$
\{\omega_1,\omega_2\}
$$

and

$$
\{\omega_3,\omega_4\}.
$$

$$

# 10) Entropy, Conditional Entropy, and Mutual Information

We use base-2 logarithms, so the unit is bits.

## a) Entropy $H(X)$

From the sample space,

$$
P(X=0)=0.50+0.20=0.70
$$

and

$$
P(X=1)=0.10+0.20=0.30.
$$

Therefore,

$$
H(X)
=
-\sum_x P(X=x)\log_2 P(X=x)
$$

$$
H(X)
=
-\left[0.70\log_2(0.70)+0.30\log_2(0.30)\right].
$$

$$
H(X)\approx 0.8813 \text{ bits}.
$$

---

## b) Conditional Entropy $H(X\mid Y)$

From earlier,

$$
P(Y=0)=0.60
$$

and

$$
P(Y=1)=0.40.
$$

For $Y=0$:

$$
P(X=0\mid Y=0)=\frac{5}{6},
\quad
P(X=1\mid Y=0)=\frac{1}{6}.
$$

So,

$$
H(X\mid Y=0)
=
-\left[
\frac{5}{6}\log_2\left(\frac{5}{6}\right)
+
\frac{1}{6}\log_2\left(\frac{1}{6}\right)
\right].
$$

$$
H(X\mid Y=0)\approx 0.6500 \text{ bits}.
$$

For $Y=1$:

$$
P(X=0\mid Y=1)=\frac{1}{2},
\quad
P(X=1\mid Y=1)=\frac{1}{2}.
$$

So,

$$
H(X\mid Y=1)=1 \text{ bit}.
$$

Therefore,

$$
H(X\mid Y)
=
P(Y=0)H(X\mid Y=0)
+
P(Y=1)H(X\mid Y=1).
$$

$$
H(X\mid Y)
=
0.60(0.6500)+0.40(1).
$$

$$
H(X\mid Y)
\approx 0.7900 \text{ bits}.
$$

---

## c) Mutual Information $I(X;Y)$

Mutual information is

$$
I(X;Y)=H(X)-H(X\mid Y).
$$

$$
I(X;Y)
=
0.8813-0.7900.
$$

$$
I(X;Y)
\approx 0.0913 \text{ bits}.
$$

---

## d) Percentage of Total Uncertainty Removed

The percentage of uncertainty removed is

$$
\frac{I(X;Y)}{H(X)}\times 100\%.
$$

$$
=
\frac{0.0913}{0.8813}\times 100\%.
$$

$$
\approx 10.36\%.
$$

Therefore, knowing the customer's risk class removes about

$$

$$

of the bank's total uncertainty about fraud.

# Q2) Conditional Density of the Multivariate Gaussian

Let

$$
Z=
\begin{bmatrix}
X\\
Y
\end{bmatrix}
\sim
\mathcal{N}
\left(
\begin{bmatrix}
\mu_X\\
\mu_Y
\end{bmatrix},
\begin{bmatrix}
\Sigma_{XX} & \Sigma_{XY}\\
\Sigma_{YX} & \Sigma_{YY}
\end{bmatrix}
\right).
$$

The conditional density is defined by

$$
f_{X\mid Y}(x\mid y)
=
\frac{f_{X,Y}(x,y)}{f_Y(y)}.
$$

Since the marginal distribution of $Y$ is

$$
Y\sim \mathcal{N}(\mu_Y,\Sigma_{YY}),
$$

we have

$$
f_Y(y)
=
\frac{1}{\sqrt{(2\pi)^m|\Sigma_{YY}|}}
\exp
\left(
-\frac{1}{2}
(y-\mu_Y)^T\Sigma_{YY}^{-1}(y-\mu_Y)
\right).
$$

The joint density is

$$
f_{X,Y}(x,y)
=
\frac{1}{\sqrt{(2\pi)^{n+m}|\Sigma|}}
\exp
\left(
-\frac{1}{2}
\begin{bmatrix}
x-\mu_X\\
y-\mu_Y
\end{bmatrix}^T
\Sigma^{-1}
\begin{bmatrix}
x-\mu_X\\
y-\mu_Y
\end{bmatrix}
\right).
$$

Therefore,

$$
f_{X\mid Y}(x\mid y)
=
\frac{f_{X,Y}(x,y)}{f_Y(y)}.
$$

After simplifying the quadratic form using block matrix inversion, the conditional distribution of $X$ given $Y=y$ is Gaussian with mean

$$
\mu^*
=
\mu_X+\Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y)
$$

and covariance matrix

$$
\Sigma_{X\mid Y}
=
\Sigma_{XX}-\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}.
$$

Hence,

$$
X\mid Y=y
\sim
\mathcal{N}
\left(
\mu^*,
\Sigma_{X\mid Y}
\right).
$$

Therefore, the conditional density is

$$
f_{X\mid Y}(x\mid y)
=
\frac{1}{\sqrt{(2\pi)^n|\Sigma_{X\mid Y}|}}
\exp
\left(
-\frac{1}{2}
(x-\mu^*)^T
\Sigma_{X\mid Y}^{-1}
(x-\mu^*)
\right).
$$

where

$$
\Sigma_{X\mid Y}
=
\Sigma_{XX}-\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}
$$

and

$$
\mu^*
=
\mu_X+\Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y).
$$

#Q3) Continuous Fault Monitoring of a Spindle Motor
## 1) Joint Distribution of $(X,Y)$

We are given

$$
X\sim \mathcal{N}(0,1)
$$

and

$$
Q\sim \mathcal{N}(0,4),
$$

with $X$ and $Q$ independent.

The measurement is

$$
Y=X+Q.
$$

Since a linear combination of independent Gaussian random variables is also Gaussian,

$$
Y\sim \mathcal{N}(\mu_Y,\sigma_Y^2).
$$

## Mean of $Y$

$$
E[Y]
=
E[X+Q]
$$

$$
=
E[X]+E[Q]
$$

$$
=
0+0
=
0.
$$

## Variance of $Y$

Because $X$ and $Q$ are independent,

$$
\operatorname{Var}(Y)
=
\operatorname{Var}(X+Q)
$$

$$
=
\operatorname{Var}(X)+\operatorname{Var}(Q)
$$

$$
=
1+4
=
5.
$$

Therefore,

$$
Y\sim \mathcal{N}(0,5).
$$

---

## Covariance $\operatorname{Cov}(X,Y)$

Using

$$
Y=X+Q,
$$

we get

$$
\operatorname{Cov}(X,Y)
=
\operatorname{Cov}(X,X+Q).
$$

Using linearity of covariance,

$$
\operatorname{Cov}(X,Y)
=
\operatorname{Cov}(X,X)
+
\operatorname{Cov}(X,Q).
$$

Since $X$ and $Q$ are independent,

$$
\operatorname{Cov}(X,Q)=0.
$$

Also,

$$
\operatorname{Cov}(X,X)=\operatorname{Var}(X)=1.
$$

Hence,

$$
\operatorname{Cov}(X,Y)=1.
$$

---

Therefore, the joint random vector is Gaussian:

$$
\begin{bmatrix}
X\\
Y
\end{bmatrix}
\sim
\mathcal{N}
\left(
\begin{bmatrix}
0\\
0
\end{bmatrix},
\begin{bmatrix}
1 & 1\\
1 & 5
\end{bmatrix}
\right).
$$

---
# 2) Information Geometry

The sigma-algebra $\sigma(Y)$ contains only the information provided by the measurement $Y$.

Here,

$$
Y=X+Q,
$$

where

$$
X\sim \mathcal{N}(0,1)
$$

and

$$
Q\sim \mathcal{N}(0,4)
$$

are independent.

Since the engineer only observes $Y$, they do not separately observe $X$ and $Q$.

Therefore, many different pairs $(X,Q)$ can produce the same value of $Y$.

For example, if

$$
Y=5,
$$

this could happen because:

- the motor temperature is high:

$$
X=5,\quad Q=0,
$$

or because

$$
X=3,\quad Q=2,
$$

or because

$$
X=1,\quad Q=4.
$$

All these combinations satisfy

$$
Y=X+Q=5.
$$

Thus, the value of $Y$ alone does not uniquely determine $X$.

Therefore, $X$ cannot be written as a function of $Y$ only, meaning $X$ is not $\sigma(Y)$-measurable.

Practically, when the power meter shows a large reading such as

$$
Y=5,
$$

the engineer cannot know with certainty whether:

- the spindle is overheating (large $X$), or
- the cutting torque/noise is large (large $Q$).

The measurement mixes both effects together, so some uncertainty about the true internal temperature always remains.

---
# 3) Conditional Expectation (The Best Estimate)

For jointly Gaussian random variables,

$$
E[X\mid Y=y]
=
\mu_X
+
\frac{\operatorname{Cov}(X,Y)}{\operatorname{Var}(Y)}
(y-\mu_Y).
$$

From Part 1:

$$
\mu_X=0,
\quad
\mu_Y=0,
$$

$$
\operatorname{Cov}(X,Y)=1,
$$

and

$$
\operatorname{Var}(Y)=5.
$$

Substituting into the formula:

$$
E[X\mid Y=y]
=
0+\frac{1}{5}(y-0).
$$

Therefore,

$$
E[X\mid Y=y]
=
\frac{1}{5}y.
$$

Hence,

$$
E[X\mid Y]
=
\frac{1}{5}Y.
$$

---

The coefficient

$$
\frac{\operatorname{Cov}(X,Y)}{\operatorname{Var}(Y)}
=
\frac{1}{5}
$$

is called the signal-to-total-variance ratio.

Here:

- Signal variance:

$$
\operatorname{Var}(X)=1
$$

- Total observed variance:

$$
\operatorname{Var}(Y)=1+4=5.
$$

Thus,

$$
\frac{1}{5}
=
\frac{\text{signal variance}}{\text{total variance}}.
$$

This means only

$$
20\%
$$

of the variability in the measurement $Y$ comes from the true temperature signal $X$, while the remaining

$$
80\%
$$

comes from the mechanical load noise $Q$.

Therefore, the best estimate of the hidden temperature is

$$
E[X\mid Y]=\frac{1}{5}Y.
$$

# 4) Verification of the Residual

Let

$$
\hat{X}=E[X\mid Y].
$$

From Part 3,

$$
\hat{X}=\frac{1}{5}Y.
$$

Define the residual error as

$$
Z=X-\hat{X}.
$$

Therefore,

$$
Z=X-\frac{1}{5}Y.
$$

Since

$$
Y=X+Q,
$$

we get

$$
Z=X-\frac{1}{5}(X+Q).
$$

$$
Z=\frac{4}{5}X-\frac{1}{5}Q.
$$

## Prove that $E[Z]=0$

$$
E[Z]
=
E\left[\frac{4}{5}X-\frac{1}{5}Q\right].
$$

$$
E[Z]
=
\frac{4}{5}E[X]-\frac{1}{5}E[Q].
$$

Since

$$
E[X]=0
$$

and

$$
E[Q]=0,
$$

we get

$$
E[Z]
=
\frac{4}{5}(0)-\frac{1}{5}(0)
=
0.
$$

Hence,

$$
E[Z]=0.
$$

## Verify that $Z$ is uncorrelated with $Y$

We need to show that

$$
\operatorname{Cov}(Z,Y)=0.
$$

Since

$$
Z=X-\frac{1}{5}Y,
$$

we have

$$
\operatorname{Cov}(Z,Y)
=
\operatorname{Cov}\left(X-\frac{1}{5}Y,Y\right).
$$

Using linearity of covariance,

$$
\operatorname{Cov}(Z,Y)
=
\operatorname{Cov}(X,Y)
-
\frac{1}{5}\operatorname{Cov}(Y,Y).
$$

Now,

$$
\operatorname{Cov}(X,Y)=1
$$

and

$$
\operatorname{Cov}(Y,Y)=\operatorname{Var}(Y)=5.
$$

Therefore,

$$
\operatorname{Cov}(Z,Y)
=
1-\frac{1}{5}(5).
$$

$$
\operatorname{Cov}(Z,Y)=1-1=0.
$$

Hence,

$$
\operatorname{Cov}(Z,Y)=0.
$$

Therefore, the residual error $Z$ is uncorrelated with the observation $Y$.

This verifies the Orthogonality Principle.

# 5) Information Theory: Differential Entropy

For a normal random variable,

$$
h(X)=\frac{1}{2}\ln(2\pi e\sigma_X^2).
$$

Here,

$$
X\sim \mathcal{N}(0,1),
$$

so

$$
\sigma_X^2=1.
$$

Therefore,

$$
h(X)
=
\frac{1}{2}\ln(2\pi e).
$$

$$
h(X)\approx 1.419 \text{ nats}.
$$

---

## Conditional Differential Entropy $h(X\mid Y)$

The conditional variance is given by

$$
\sigma_{X\mid Y}^2
=
\sigma_X^2
-
\frac{\operatorname{Cov}(X,Y)^2}{\operatorname{Var}(Y)}.
$$

From earlier,

$$
\sigma_X^2=1,
\quad
\operatorname{Cov}(X,Y)=1,
\quad
\operatorname{Var}(Y)=5.
$$

So,

$$
\sigma_{X\mid Y}^2
=
1-\frac{1^2}{5}.
$$

$$
\sigma_{X\mid Y}^2
=
1-\frac{1}{5}
=
\frac{4}{5}.
$$

Therefore,

$$
h(X\mid Y)
=
\frac{1}{2}\ln(2\pi e\sigma_{X\mid Y}^2).
$$

$$
h(X\mid Y)
=
\frac{1}{2}\ln\left(2\pi e\cdot \frac{4}{5}\right).
$$

$$
h(X\mid Y)
\approx 1.307 \text{ nats}.
$$

---
# 6) Mutual Information

The mutual information is

$$
I(X;Y)=h(X)-h(X\mid Y).
$$

From Part 5,

$$
h(X)=\frac{1}{2}\ln(2\pi e)
$$

and

$$
h(X\mid Y)
=
\frac{1}{2}\ln\left(2\pi e\cdot \frac{4}{5}\right).
$$

Therefore,

$$
I(X;Y)
=
\frac{1}{2}\ln(2\pi e)
-
\frac{1}{2}\ln\left(2\pi e\cdot \frac{4}{5}\right).
$$

Using logarithm properties,

$$
I(X;Y)
=
\frac{1}{2}\ln\left(\frac{2\pi e}{2\pi e\cdot \frac{4}{5}}\right).
$$

$$
I(X;Y)
=
\frac{1}{2}\ln\left(\frac{5}{4}\right).
$$

$$
I(X;Y)\approx 0.1116 \text{ nats}.
$$

---

## Percentage of Uncertainty Removed

The percentage of uncertainty removed is

$$
\frac{I(X;Y)}{h(X)}\times 100\%.
$$

Substituting values,

$$
=
\frac{0.1116}{1.419}\times 100\%.
$$

$$
\approx 7.87\%.
$$

Therefore, observing the power meter removes approximately

$$
7.87\%
$$

of the uncertainty about the hidden temperature.

# Effect of Larger Mechanical Noise

Suppose now the mechanical noise has variance

$$
\operatorname{Var}(Q)=100.
$$

Then

$$
Y=X+Q
$$

has variance

$$
\operatorname{Var}(Y)=1+100=101.
$$

Also,

$$
\operatorname{Cov}(X,Y)=1.
$$

The conditional variance becomes

$$
\sigma_{X\mid Y}^2
=
1-\frac{1^2}{101}.
$$

$$
=
\frac{100}{101}.
$$

Thus,

$$
I(X;Y)
=
\frac{1}{2}\ln\left(\frac{1}{100/101}\right).
$$

$$
I(X;Y)
=
\frac{1}{2}\ln\left(\frac{101}{100}\right).
$$

$$
I(X;Y)\approx 0.00498 \text{ nats}.
$$

This is much smaller than before.

Practically, when the mechanical noise becomes extremely large, the observation $Y$ is dominated by noise, so the power meter provides almost no useful information about the true internal temperature $X$.

# 7) Diagnostic Interpretation

The manager says that since

$$
Y=X+Q,
$$

the best estimate of temperature should be

$$
Y
$$

itself.

But this is not correct because $Y$ contains both the true temperature $X$ and the mechanical noise $Q$.

From Question 3,

$$
E[X\mid Y]=\frac{1}{5}Y.
$$

This means the best estimate is not $Y$, but only one-fifth of $Y$.

If the power meter shows

$$
Y=5,
$$

then the best estimate of temperature is

$$
E[X\mid Y=5]=\frac{1}{5}(5)=1.
$$

So using $Y=5$ directly would overestimate the true temperature.

The estimator $Y$ is biased for $X$ after observing $Y$ because

$$
E[X\mid Y=y]=\frac{1}{5}y
$$

but the manager's estimate is

$$
y.
$$

The conditional bias is

$$
y-\frac{1}{5}y=\frac{4}{5}y.
$$

Therefore, $Y$ is pulled too far away from the mean.

Since most of the variation in $Y$ comes from noise $Q$, the best estimate must shrink the measurement toward the mean zero.

That is why

$$
E[X\mid Y]=\frac{1}{5}Y
$$

is a shrinkage estimator.
