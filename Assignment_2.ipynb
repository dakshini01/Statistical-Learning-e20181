# Q3) Transformer Failures

The number of emergency transformer failures in a single day is modeled as a discrete random variable $X$, where

$$
X \in \{0,1,2,3\}
$$

with probabilities

$$
P(X=0)=0.50,\quad P(X=1)=0.30,\quad P(X=2)=0.15,\quad P(X=3)=0.05
$$

---

## 1. Probability Space

A suitable probability space is

$$
\Omega=\{0,1,2,3\}
$$

$$
\mathcal{F}=2^\Omega
$$

with probability measure

$$
P(\{0\})=0.50
$$

$$
P(\{1\})=0.30
$$

$$
P(\{2\})=0.15
$$

$$
P(\{3\})=0.05
$$

Thus,

$$
(\Omega,\mathcal{F},P)
$$

is the probability space.

---

## 2. Random Variable

Define the random variable

$$
X:\Omega \rightarrow \{0,1,2,3\}
$$

by

$$
X(\omega)=\omega
$$

Thus, $X$ directly represents the number of transformer failures occurring in one day.

---

## 3. Probability Mass Function (PMF)

The PMF of $X$ is

$$
p_X(x)=P(X=x)
$$

$$
p_X(x)=
\begin{cases}
0.50, & x=0 \\
0.30, & x=1 \\
0.15, & x=2 \\
0.05, & x=3 \\
0, & \text{otherwise}
\end{cases}
$$

Check normalization:

$$
0.50+0.30+0.15+0.05=1
$$

Hence, this is a valid PMF.

---

## 4. Expectation

The expectation is

$$
E[X]=\sum_x xP(X=x)
$$

Therefore,

$$
E[X]=(0)(0.50)+(1)(0.30)+(2)(0.15)+(3)(0.05)
$$

$$
E[X]=0+0.30+0.30+0.15
$$

$$
E[X]=0.75
$$

Hence,

$$
\boxed{E[X]=0.75}
$$

---

## 5. Variance

First compute $E[X^2]$:

$$
E[X^2]=\sum_x x^2P(X=x)
$$

$$
E[X^2]=(0^2)(0.50)+(1^2)(0.30)+(2^2)(0.15)+(3^2)(0.05)
$$

$$
E[X^2]=0+0.30+0.60+0.45
$$

$$
E[X^2]=1.35
$$

Now,

$$
Var(X)=E[X^2]-(E[X])^2
$$

$$
Var(X)=1.35-(0.75)^2
$$

$$
Var(X)=1.35-0.5625
$$

$$
\boxed{Var(X)=0.7875}
$$

---

## 6. Required Probabilities

### Probability of at least one failure

$$
P(X\ge1)=P(X=1)+P(X=2)+P(X=3)
$$

$$
P(X\ge1)=0.30+0.15+0.05
$$

$$
\boxed{P(X\ge1)=0.50}
$$

---

### Probability of at least two failures

$$
P(X\ge2)=P(X=2)+P(X=3)
$$

$$
P(X\ge2)=0.15+0.05
$$

$$
\boxed{P(X\ge2)=0.20}
$$

---

## 7. Practical Interpretation

The expectation

$$
E[X]=0.75
$$

means that the district experiences an average of $0.75$ transformer failures per day over the long run.

The variance

$$
Var(X)=0.7875
$$

measures how much the daily number of transformer failures fluctuates around the average value. This indicates a moderate level of variability in transformer failures.

# Q4) Probability of Defective Items

The defective proportion is modeled as a continuous random variable $X$ such that

$$
X \in [0,1]
$$

and

$$
X \sim B(2,18)
$$

---

## 1. Probability Space

A suitable probability space is

$$
\Omega=[0,1]
$$

$$
\mathcal{F}=\mathcal{B}([0,1])
$$

where $\mathcal{B}([0,1])$ is the Borel sigma-algebra on $[0,1]$.

The probability measure $P$ is defined by the density function

$$
P(A)=\int_A f(x)\,dx
$$

for any measurable set $A \in \mathcal{F}$.

Thus, the probability space is

$$
(\Omega,\mathcal{F},P)
$$

---

## 2. Random Variable

Define

$$
X:\Omega \rightarrow [0,1]
$$

by

$$
X(\omega)=\omega
$$

Here, $X$ represents the proportion of defective items in the batch.

---

## 3. Probability Density Function

The probability density function (PDF) of $X$ is

The probability density function of $X$ is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1 \\
0, & \text{otherwise}
\end{cases}
$$

---

## 4. Verification of Valid PDF

For $0<x<1$,

$$
x^{1}(1-x)^{17} \ge 0
$$

and

$$
B(2,18)>0
$$

Therefore,

$$
f(x)\ge 0
$$

Also,

$$
\int_{-\infty}^{\infty} f(x)\,dx
=
\int_0^1 \frac{1}{B(2,18)}x^{1}(1-x)^{17}\,dx
$$

By definition of the Beta function,

$$
B(2,18)=\int_0^1 x^{1}(1-x)^{17}\,dx
$$

Therefore,

$$
\int_0^1 \frac{1}{B(2,18)}x^{1}(1-x)^{17}\,dx=1
$$

Hence, $f(x)$ is a valid probability density function.

---

## 5. Expectation

The expectation of a continuous random variable is defined by

$$
E[X]=\int_{-\infty}^{\infty} x f_X(x)\,dx
$$

Since the support of the Beta distribution is $(0,1)$,

$$
E[X]=\int_0^1 x f_X(x)\,dx
$$

Substituting the PDF,

$$
E[X]
=
\int_0^1
x
\left(
\frac{1}{B(\alpha,\beta)}
x^{\alpha-1}(1-x)^{\beta-1}
\right)
dx
$$

$$
E[X]
=
\frac{1}{B(\alpha,\beta)}
\int_0^1
x^\alpha(1-x)^{\beta-1}
dx
$$

Using the Beta function identity

$$
B(m,n)=\int_0^1 x^{m-1}(1-x)^{n-1}dx
$$

we obtain

$$
\int_0^1 x^\alpha(1-x)^{\beta-1}dx
=
B(\alpha+1,\beta)
$$

Therefore,

$$
E[X]
=
\frac{B(\alpha+1,\beta)}{B(\alpha,\beta)}
$$

Using the Beta-function property,

$$
\frac{B(\alpha+1,\beta)}{B(\alpha,\beta)}
=
\frac{\alpha}{\alpha+\beta}
$$

Thus,

$$
E[X]=\frac{\alpha}{\alpha+\beta}
$$

Substituting $\alpha=2$ and $\beta=18$,

$$
E[X]=\frac{2}{2+18}
$$

$$
E[X]=\frac{2}{20}
$$

$$
\boxed{E[X]=0.10}
$$

---

## 6. Variance

The variance is defined by

$$
Var(X)=E[X^2]-(E[X])^2
$$

First compute $E[X^2]$:

$$
E[X^2]
=
\int_0^1 x^2 f_X(x)\,dx
$$

Substituting the PDF,

$$
E[X^2]
=
\int_0^1
x^2
\left(
\frac{1}{B(\alpha,\beta)}
x^{\alpha-1}(1-x)^{\beta-1}
\right)
dx
$$

$$
E[X^2]
=
\frac{1}{B(\alpha,\beta)}
\int_0^1
x^{\alpha+1}(1-x)^{\beta-1}dx
$$

Using the Beta function identity again,

$$
\int_0^1
x^{\alpha+1}(1-x)^{\beta-1}dx
=
B(\alpha+2,\beta)
$$

Therefore,

$$
E[X^2]
=
\frac{B(\alpha+2,\beta)}{B(\alpha,\beta)}
$$

Using Beta-function properties,

$$
E[X^2]
=
\frac{\alpha(\alpha+1)}
{(\alpha+\beta)(\alpha+\beta+1)}
$$

Now,

$$
Var(X)=E[X^2]-(E[X])^2
$$

Substituting the expressions,

$$
Var(X)
=
\frac{\alpha(\alpha+1)}
{(\alpha+\beta)(\alpha+\beta+1)}
-
\left(
\frac{\alpha}{\alpha+\beta}
\right)^2
$$

Simplifying,

$$
Var(X)
=
\frac{\alpha\beta}
{(\alpha+\beta)^2(\alpha+\beta+1)}
$$

Substituting $\alpha=2$ and $\beta=18$,

$$
Var(X)
=
\frac{(2)(18)}
{(2+18)^2(2+18+1)}
$$

$$
Var(X)
=
\frac{36}{(20)^2(21)}
$$

$$
Var(X)
=
\frac{36}{8400}
$$

$$
\boxed{Var(X)=0.004286}
$$

A Beta distribution is suitable because the defective proportion must lie between $0$ and $1$. Since $X$ represents a proportion, it cannot be negative or greater than $1$.

The Beta distribution is flexible and can model different shapes depending on the parameters $\alpha$ and $\beta$. In this case, $\alpha=2$ and $\beta=18$, so the distribution is concentrated near smaller values. This is reasonable for a manufacturing process where the defective proportion is expected to be low.

## 9. Why the PERT Distribution May Be More Appropriate than a Triangular Distribution

The PERT distribution may be more appropriate because it provides a smoother and more realistic representation of uncertainty in project completion times.

Unlike the Triangular distribution, the PERT distribution is based on a scaled Beta distribution, which produces a smoother probability density curve rather than sharp corners. This makes the model less sensitive to extreme values and gives more emphasis to the most likely completion time.

The expectation of the PERT distribution is

$$
E[X]=\frac{a+4c+b}{6}
$$

which assigns four times more weight to the most likely estimate $c$. This reflects practical project situations where the completion time is expected to occur closer to the most likely value than to the optimistic or pessimistic extremes.

In contrast, the Triangular distribution treats the distribution more linearly and may overemphasize the boundary values.

Therefore, the PERT distribution is often preferred in project management because it provides a more realistic estimate of project duration and uncertainty.
